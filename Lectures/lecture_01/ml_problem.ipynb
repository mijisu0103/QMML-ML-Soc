## Simple Implementation of a Gradient Descent Algorithm ##

In this notebook, we will implement a basic version of the gradient descent algorithm to minimize a simple mathematical function. The process will involve the following steps:

1. **Define the function**: We will define a function \( f(x) \) that we aim to minimize.
2. **Visualize the function**: We will plot the function to understand its behavior.
3. **Implement gradient descent**: We will compute the gradient (derivative) of the function and iteratively update our parameter \( x \) to minimize \( f(x) \).
4. **Plot the results**: Finally, we will plot the path taken by gradient descent to find the minimum.

This will provide a simple and intuitive way to understand how gradient descent works in practice.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define our loss function

def f(x):
    return (x-2)**2 + 1 / 2 * np.sin(x)

# Define initial point

p0 = 0

# Return f(x) evaluated at p0

print(f"f(x) evaluated at x = 0 is {f(p0)}.")
                                
```

```python
# Define the derivative of the loss function

def f_d1(x):
    return 2*(x-2) + 1 / 2 * np.cos(x)
```

```python
x = np.linspace(-1,5,400)
y = f(x)

plt.plot(x,y, label = 'f(x)')
plt.scatter(p0, f(p0), color='red', zorder=5, label=f'Point at x={x}')
plt.title("Graph of f(x) and f(0)")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.show()
```

```python
def gradient_descent(initial_x,learning_rate, epochs):
    
    """This function finds the minimum value of f via 
    the gradient descent algorithm."""

    i = 0
    x = initial_x
    
    while i < epochs:
        df_x = f_d1(x)
        x = x - learning_rate * df_x
        if i % 500 == 0:
            print(f"Iteration number: {i} x = {x}, f(x) = {f(x)}")
        i += 1

    return x, f(x)
```

```python
# Plot results on the graph

x_point, y_point = gradient_descent(p0,0.001,4001)
```

```python
plt.plot(x,y, label = 'f(x)')
plt.scatter(x_point, y_point, color='red', zorder=5, label=f'Point at x={x_point}')
plt.title("Graph of f(x)=(x-2)^2+1/2*sin(x)")
plt.annotate('This is the point (2.129, 0.44074)', 
             xy=(x_point, y_point),            # Point to annotate
             xytext=(x_point-1, y_point + 5 ), # Position of the comment text
             arrowprops=dict(facecolor='black', arrowstyle='->'),  # Arrow properties
             fontsize=10)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.show()
```

```python
# Define the first function and its derivative
def f1(x):
    return (x - 3) ** 2 + 12

def f1_d1(x):
    return 2 * (x - 3)

# Define the second function and its derivative
def f2(x):
    return np.sin(x)

def f2_d1(x):
    return np.cos(x)

# Define the gradient descent function following the notebook's method
def gradient_descent(initial_x, learning_rate, epochs, f, f_d1):
    i = 0
    x = initial_x
    
    while i < epochs:
        df_x = f_d1(x)
        x = x - learning_rate * df_x
        if i % 10 == 0:  # Print at intervals of 10 for better readability
            print(f"Iteration number: {i} x = {x}, f(x) = {f(x)}")
        i += 1

    return x, f(x)

# Initial values for gradient descent
initial_x1 = 0  # Starting point for f1(x)
initial_x2 = 3  # Starting point for f2(x)

learning_rate = 0.1
epochs = 50

# Run gradient descent for both functions
print("Gradient Descent for f1(x) = (x-3)^2 + 12")
x1_min, y1_min = gradient_descent(initial_x1, learning_rate, epochs, f1, f1_d1)

print("\nGradient Descent for f2(x) = sin(x)")
x2_min, y2_min = gradient_descent(initial_x2, learning_rate, epochs, f2, f2_d1)
```

```python
# Generate x values for plotting
x_range = np.linspace(-5, 8, 400)

# Gradient descent results for both functions
x1_point, y1_point = x1_min, y1_min
x2_point, y2_point = x2_min, y2_min

# Plot for f1(x) = (x-3)^2 + 12
plt.figure(figsize=(10, 4))
plt.plot(x_range, f1(x_range), label=r"$f(x) = (x-3)^2 + 12$")
plt.scatter(x1_point, y1_point, color='red', zorder=5, label=f'Point at x={x1_point:.3f}')

plt.annotate(f'This is the point ({x1_point:.3f}, {y1_point:.3f})', 
             xy=(x1_point, y1_point),  
             xytext=(x1_point-1, y1_point + 5),  
             arrowprops=dict(facecolor='black', arrowstyle='->'),  
             fontsize=10)

plt.title("Graph of f(x) = (x-3)^2 + 12")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()

# Plot for f2(x) = sin(x)
plt.figure(figsize=(10, 4))
plt.plot(x_range, f2(x_range), label=r"$f(x) = \sin{x}$")
plt.scatter(x2_point, y2_point, color='red', zorder=5, label=f'Point at x={x2_point:.3f}')

plt.annotate(f'This is the point ({x2_point:.3f}, {y2_point:.3f})', 
             xy=(x2_point, y2_point),  
             xytext=(x2_point-1, y2_point + 0.5),  
             arrowprops=dict(facecolor='black', arrowstyle='->'),  
             fontsize=10)

plt.title("Graph of f(x) = sin(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
